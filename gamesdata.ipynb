```python
#Importing libraries
import pandas as pd
import lxml
```

```python
## 2017
# Importing data from 2017
games2017 = pd.read_html('prosjekt\\2017/games.xls', encoding='utf-8')
rank_table2017 = pd.read_html('prosjekt\\2017/table.xls', encoding='utf-8')
team_stats2017 = pd.read_html('prosjekt\\2017/team-stats.xls', encoding='utf-8')
player_stats2017 = pd.read_html('prosjekt\\2017/player-stats.xls', encoding='utf-8')

# Read_html returns a list, converting to dataframe and sorting by selected values
games2017 = games2017[0]
games2017 = games2017[games2017['Wk'].notnull()]
rank_table2017 = rank_table2017[0]
team_stats2017 = team_stats2017[0]
player_stats2017 = player_stats2017[0]
```

```python
## 2018
#Importing data from 2018
games2018 = pd.read_html('prosjekt\\2018/games.xls', encoding='utf-8')
rank_table2018 = pd.read_html('prosjekt\\2018/table.xls', encoding='utf-8')
team_stats2018 = pd.read_html('prosjekt\\2018/team-stats.xls', encoding='utf-8')
player_stats2018 = pd.read_html('prosjekt\\2018/player-stats.xls', encoding='utf-8')

# Read_html returns a list, converting to dataframe and sorting by certain values
games2018 = games2018[0]
games2018 = games2018[games2018['Wk'].notnull()]
rank_table2018 = rank_table2018[0]
team_stats2018 = team_stats2018[0]
player_stats2018 = player_stats2018[0]
```

```python
## 2019
#Importing data from 2019
games2019 = pd.read_html('prosjekt\\2019/games.xls', encoding='utf-8')
rank_table2019 = pd.read_html('prosjekt\\2019/table.xls', encoding='utf-8')
team_stats2019 = pd.read_html('prosjekt\\2019/team-stats.xls', encoding='utf-8')
player_stats2019 = pd.read_html('prosjekt\\2019/player-stats.xls', encoding='utf-8')

# read_html returns a list, one df for every table in the document
games2019 = games2019[0]
games2019 = games2019[games2019['Wk'].notnull()]
rank_table2019 = rank_table2019[0]
team_stats2019 = team_stats2019[0]
player_stats2019 = player_stats2019[0]
```

```python
## 2017
# Splitting the 'Score' column into HomeScore and AwayScore
home, away = games2017['Score'].str.split("–").str

# Preparing to select columns from team-stats file that we are interested in
perf_fields = ['Gls', 'PK', 'PKatt', 'CrdY', 'CrdR']
per_90_fields = ['Gls', 'G-PK']

# Concatenating columns
team_stats2017 = pd.concat([
    team_stats2017['Unnamed: 0_level_0'][['Squad']],
    team_stats2017['Unnamed: 2_level_0'][['Age']].astype(float), 
    team_stats2017['Performance'][perf_fields].astype(int), 
    team_stats2017['Per 90 Minutes'][per_90_fields].add_prefix('90-').astype(float)
], axis=1).sort_values(by=['Squad'], ascending=True)

# Adding columns 'Rk' and 'GF'
team_stats2017['Rk'] = rank_table2017['Rk']
team_stats2017['GF'] = rank_table2017['GF']

team_stats2017.head(3)
```

```python
## 2017
# Converting columns into type int
games2017['HomeScore'] = home.astype(int)
games2017['AwayScore'] = away.astype(int)

# Grouping by 'Home' to then calculate mean of 'HomeScore' and 'AwayScore'
games_by_home_team2017 = games2017.groupby("Home")
home_team_averages2017 = games_by_home_team2017.mean().reset_index()
home_team_averages2017 = home_team_averages2017[['Home','HomeScore', 'AwayScore']]

# Renaming columns
home_team_averages2017 = home_team_averages2017.rename(columns={'Home': 'Squad'})
home_team_averages2017 = home_team_averages2017.rename(columns={'HomeScore': 'HomeScoreMean'})
home_team_averages2017 = home_team_averages2017.rename(columns={'AwayScore': 'AwayScoreMean'})

home_team_averages2017.head(3)
```

```python
## 2017
# Quick look at the games2017 dataframe
games2017.head(3)
```

```python
## 2017
# Selecting only necessary columns
games2017 = games2017[["Home", "Away", "HomeScore", "AwayScore"]]
games2017.head(3)
```

```python
## 2017
# Merging games2017 with home_team_averages2017 into a new dataframe called games2017_home
games2017_home = pd.merge(games2017, home_team_averages2017, how='left', left_on='Home', right_on='Squad')
games2017_home = games2017_home.drop(columns=['Squad'])

# Merging again with team_stats2017
games2017_home = pd.merge(games2017_home, team_stats2017, how='left', left_on='Home', right_on='Squad')

#Adding an extra column 'IsHome' that will be used later
games2017_home['IsHome'] = pd.Series([1 for x in range(len(games2017.index))])

games2017_home.head(3)
```

```python
## 2017
# Merging games2017 with home_team_averages2017 into a new dataframe called games2017_away
games2017_away = pd.merge(games2017, home_team_averages2017, how='left', left_on='Away', right_on='Squad')
games2017_away = games2017_away.drop(columns=['Squad'])

# Merging again with team_stats2017
games2017_away = pd.merge(games2017_away, team_stats2017, how='left', left_on='Away', right_on='Squad')

#Adding an extra column 'IsHome' that will be used later
games2017_away['IsHome'] = pd.Series([0 for x in range(len(games2017.index))])

games2017_away
```

```python
## 2017
# Concatenating the two dataframes
games2017 = pd.concat([games2017_home, games2017_away]).sort_index()
games2017 = games2017.drop('Squad', axis=1).reset_index().drop(['index'], axis=1)

# Adding another column with year of dataset
games2017['Year'] = pd.Series([2017 for x in range(len(games2017.index))])
```
